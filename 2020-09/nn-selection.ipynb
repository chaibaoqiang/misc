```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider the following dataset with Bernoulli-distributed random variables: `D = [0, 1, 1, 1, 0, 1, 1, 1]`. Since the probability mass function (pmf) or a bernoulli-distributed random variable is given by

$$
    p(x\vert\mu) = \mu^x (1 - \mu)^{1-x},
$$

we can see that our likelihood will take the form

$$
    \prod_{n=1}^8 p(x_n\vert\mu) = \mu^{\sum_{n=1}^8x_n} (1 - \mu)^{N - \sum_{n=1}^8x_n}
$$

```python
D = np.array([0, 1, 1, 1, 0, 1, 1, 1])
N = len(D)

mu = np.linspace(0, 1, 100)
L = mu ** (D.sum()) * (1 - mu) ** (N - D.sum())
log_L = np.log(L)

fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.plot(mu, L)
ax2.plot(mu, log_L, color="tab:orange")
plt.title("Likelihood of $D$");
ax.set_xlabel(r"$\mu$", fontsize=13)
ax.set_ylabel(r"$\mathcal{L}$", fontsize=13);
ax2.set_ylabel(r"$\log\mathcal{L}$", fontsize=13);
plt.savefig("/Users/gerardoduran/Desktop/log_lik.png")
```
